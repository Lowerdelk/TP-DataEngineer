```python
import pandas as pd
import fastapi as app
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
def idAmazon(df):
    dataframe = pd.read_csv(df)
    for i in dataframe['show_id']:
        dataframe['show_id'] = dataframe['show_id'].replace({i: 'a' + i})
    return dataframe


def idDisney(df):
    dataframe = pd.read_csv(df)
    for i in dataframe['show_id']:
        dataframe['show_id'] = dataframe['show_id'].replace({i: 'd' + i})
    return dataframe


def idHulu(df):
    dataframe = pd.read_csv(df)
    for i in dataframe['show_id']:
        dataframe['show_id'] = dataframe['show_id'].replace({i: 'h' + i})
    return dataframe


def idNetflix(df):
    dataframe = pd.read_csv(df)
    for i in dataframe['show_id']:
        dataframe['show_id'] = dataframe['show_id'].replace({i: 'n' + i})
    return dataframe


def nulaG(df):
    dataframe = df
    dataframe["rating"] = dataframe["rating"].fillna('g')
    return dataframe


def limpiarfechas(df):
    dataframe = df
    dataframe["date_added"] = pd.to_datetime(dataframe["date_added"])
    return dataframe


def mayuscula(df):
    dataframe = df.apply(lambda x: x.astype(str).str.lower() if(x.dtype == 'object') else x)
    return dataframe


def durationSplit(df):
    dataframe = df
    dataframe = dataframe[dataframe["duration"].notna()]
    new = dataframe["duration"].str.split(" ", n=1, expand=True)
    dataframe["duration_int"] = new[0].astype(int)
    dataframe["duration_type"] = new[1].astype(str)
    dataframe.drop(columns=["duration"], inplace=True)
    return dataframe
```

```python
amazon = idAmazon("./Datasets/amazon_prime_titles-score.csv")
disney = idDisney("./Datasets/disney_plus_titles-score.csv")
hulu = idHulu("./Datasets/hulu_titles-score (2).csv")
netflix = idNetflix("./Datasets/netflix_titles-score.csv")
```

```python
durationSplit(amazon)
durationSplit(disney)
durationSplit(hulu)
durationSplit(netflix)
```

```python
nulaG(amazon)
nulaG(disney)
nulaG(hulu)
nulaG(netflix)
```

```python
limpiarfechas(amazon)
limpiarfechas(disney)
limpiarfechas(hulu)
limpiarfechas(netflix)
```

```python
mayuscula(amazon)
mayuscula(disney)
mayuscula(hulu)
mayuscula(netflix)
```
