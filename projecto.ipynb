```python
import pandas as pd
import fastapi as app
```

```python
def idAmazon(df):
    dataframe = pd.read_csv(df)
    for i in dataframe['show_id']:
        dataframe['show_id'] = dataframe['show_id'].replace({i: 'a' + i})
    return dataframe


def idDisney(df):
    dataframe = pd.read_csv(df)
    for i in dataframe['show_id']:
        dataframe['show_id'] = dataframe['show_id'].replace({i: 'd' + i})
    return dataframe


def idHulu(df):
    dataframe = pd.read_csv(df)
    for i in dataframe['show_id']:
        dataframe['show_id'] = dataframe['show_id'].replace({i: 'h' + i})
    return dataframe


def idNetflix(df):
    dataframe = pd.read_csv(df)
    for i in dataframe['show_id']:
        dataframe['show_id'] = dataframe['show_id'].replace({i: 'n' + i})
    return dataframe


def nulaG(df):
    df["rating"] = df["rating"].fillna('g')
    return df


def limpiarfechas(df):
    df["date_added"] = pd.to_datetime(df["date_added"])


def mayuscula(df):
    df=df.apply(lambda x: x.astype(str).str.lower() if(x.dtype == 'object') else x)
    return df

def durationSplit(df):
    new = df["duration"].str.split(" ", n = 1, expand = True)
    df.drop(columns = ["duration"], inplace = True)
    df["duration_int"] = new[0].astype(int)
    df["duration_type"] = new[1].astype(str)
```

```python
amazon = idAmazon("./Datasets/amazon_prime_titles-score.csv")
```

```python
disney = idDisney("./Datasets/disney_plus_titles-score.csv")
```

```python
hulu = idHulu("./Datasets/hulu_titles-score (2).csv")
```

```python
netflix = idNetflix("./Datasets/netflix_titles-score.csv")
```

```python
nulaG(amazon)
nulaG(disney)
nulaG(hulu)
nulaG(netflix)
```

```python
limpiarfechas(amazon)
limpiarfechas(disney)
limpiarfechas(hulu)
limpiarfechas(netflix)
```

```python
mayuscula(amazon)
mayuscula(disney)
mayuscula(hulu)
mayuscula(netflix)
```
